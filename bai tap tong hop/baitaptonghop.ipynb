```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
```

```python
babi_df= pd.read_csv(r'C:\Users\Admin\Downloads\hoc sau\DuLieu\DuLieu\babies.txt')
```

```python
# Tách cột 'bwt smoke' thành hai cột riêng biệt 'bwt' và 'smoke'
babi_df[['bwt', 'smoke']] = babi_df['bwt smoke'].str.split(expand=True)

# Loại bỏ cột gốc 'bwt smoke'
babi_df = babi_df.drop(columns=['bwt smoke'])
```

```python
babi_df.info()
```

```python
# Chuyển đổi cột 'bwt' và 'smoke' về kiểu dữ liệu phù hợp
babi_df['bwt'] = babi_df['bwt'].astype(int)
babi_df['smoke'] = babi_df['smoke'].astype(int)
```

```python
# Tách tập dữ liệu nếu cột smoke bằng 0
df_smoke_0 = babi_df[babi_df['smoke'] == 0]

# Tách tập dữ liệu nếu cột smoke bằng 1
df_smoke_1 = babi_df[babi_df['smoke'] == 1]
```

```python
df_smoke_1.info()
```

```python
#Xét tập dữ liệu ứng với trƣờng hợp bà mẹ không hút thuốc
#Số lượng
count = df_smoke_0['bwt'].count()

# Min
min_val = df_smoke_0['bwt'].min()

# Max
max_val = df_smoke_0['bwt'].max()

# Mean
mean_val = df_smoke_0['bwt'].mean()

# Sd
sd_val = df_smoke_0['bwt'].std()

# Var
var_val = df_smoke_0['bwt'].var()

# Median
median_val = df_smoke_0['bwt'].median()

# Quantiles
quantile_0 = df_smoke_0['bwt'].quantile(0)
quantile_25 = df_smoke_0['bwt'].quantile(0.25)
quantile_50 = df_smoke_0['bwt'].quantile(0.5)
quantile_75 = df_smoke_0['bwt'].quantile(0.75)

# IQR (Interquartile Range)
iqr_val = quantile_75 - quantile_25

# Skewness
skewness_val = skew(df_smoke_0['bwt'])

# Kurtosis
kurtosis_val = kurtosis(df_smoke_0['bwt'])

# In kết quả
print(f"Số lượng: {count}")
print(f"Min: {min_val}")
print(f"Max: {max_val}")
print(f"Mean: {mean_val}")
print(f"Sd: {sd_val}")
print(f"Var: {var_val}")
print(f"Median: {median_val}")
print(f"Quantile 0%: {quantile_0}")
print(f"Quantile 25%: {quantile_25}")
print(f"Quantile 50%: {quantile_50}")
print(f"Quantile 75%: {quantile_75}")
print(f"IQR: {iqr_val}")
print(f"Skewness: {skewness_val}")
print(f"Kurtosis: {kurtosis_val}")

# Nhận xét
print("\nNhận xét:")
print(f"Tính biến động của dữ liệu:")
print(f"Phương sai (variance): var= {var_val}")
print(f"Độ lệch chuẩn (standard deviation): sd= {sd_val}")
print(f"Khoảng giá trị: min={min_val}, max={max_val} -> range={max_val - min_val}")
print(f"Khoảng cách giữa 2 phần tư vị: IQR=Q3-Q1={quantile_75}-{quantile_25}={iqr_val}")
print("Như vậy dữ liệu phân bố gần nhau.")
print(f"Hình dạng phân bố của dữ liệu:")
print(f"Độ lệch: Skewness={skewness_val}")
print(f"Độ bè nhọn của đỉnh dữ liệu: Kurtosis={kurtosis_val}")
```

So sánh hai tập dữ liệu dựa vào các biểu diễn hình học:

```python
# Vẽ hai biểu đồ histogram cạnh nhau
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Histogram for the first dataset
ax1.hist(df_smoke_1['bwt'], bins=10, edgecolor='black', alpha=0.7)
ax1.set_title('Histogram trong trường hợp bà mẹ có hút thuốc:')
ax1.set_xlabel('Birth Weight')
ax1.grid(True)

# Histogram for the second dataset
ax2.hist(df_smoke_0['bwt'], bins=10, edgecolor='black', alpha=0.7)
ax2.set_title('Histogram trong trường hợp bà mẹ không hút thuốc:')
ax2.set_xlabel('Birth Weight')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Vẽ boxplot cho cột 'bwt' dựa trên cột 'smoke'
babi_df.boxplot(column='bwt', by='smoke', patch_artist=True, grid=True)

plt.title('Boxplot grouped by smoke bwt')
plt.suptitle('')  # Loại bỏ tiêu đề mặc định của Pandas
plt.xlabel('Smoke')
plt.xticks([1, 2, 3], ['0', '1', '9'])  # Gắn nhãn trục x

plt.show()
```

So sánh QQ-plot phân bố của 2 tập dữ liệu

```python
import scipy.stats as stats
```

```python
# Vẽ hai QQ-plot cạnh nhau
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# QQ-plot for the first dataset
stats.probplot(df_smoke_1['bwt'], dist="norm", plot=ax1)
ax1.set_title('QQ-Plot trong trường hợp bà mẹ có hút thuốc:')
ax1.grid(True)

# QQ-plot for the second dataset
stats.probplot(df_smoke_0['bwt'], dist="norm", plot=ax2)
ax2.set_title('QQ-Plot trong trường hợp bà mẹ không hút thuốc')
ax2.grid(True)

plt.tight_layout()
plt.show()
```
