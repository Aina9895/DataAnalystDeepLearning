7. Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12. 
- Công thức tính: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10

```python
# Tính TBM1 (trung bình môn năm lớp 10)
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
```

```python
# Tính TBM2 (trung bình môn năm lớp 11) 
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
```

```python
# Tính TBM3 (trung bình môn năm lớp 12)
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10
```

```python
# Tính toán điểm trung bình tổng kết (TBT) cho mỗi học sinh:
df['TBT'] = (df['TBM1'] + df['TBM2'] + df['TBM3']) / 3
```

```python
#Tính toán điểm thi đại học trung bình (DHTB) cho mỗi học sinh:
df['DHTB'] = (df['DH1'] + df['DH2'] + df['DH3']) / 3
```

8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau:
- Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
- Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
- Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
- Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
- Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)

```python
# Phân loại năm lớp 10
df.loc[df['TBM1'] < 5, 'XL1'] = 'Y'
df.loc[(df['TBM1'] >= 5) & (df['TBM1'] < 6.5), 'XL1'] = 'TB'
df.loc[(df['TBM1'] >= 6.5) & (df['TBM1'] < 8.0), 'XL1'] = 'K'
df.loc[(df['TBM1'] >= 8.0) & (df['TBM1'] < 9.0), 'XL1'] = 'G'
df.loc[df['TBM1'] >= 9.0, 'XL1'] = 'XS'
```

```python
# Phân loại năm lớp 11 
df.loc[df['TBM2'] < 5, 'XL2'] = 'Y'
df.loc[(df['TBM2'] >= 5) & (df['TBM2'] < 6.5), 'XL2'] = 'TB'
df.loc[(df['TBM2'] >= 6.5) & (df['TBM2'] < 8.0), 'XL2'] = 'K' 
df.loc[(df['TBM2'] >= 8.0) & (df['TBM2'] < 9.0), 'XL2'] = 'G'
df.loc[df['TBM2'] >= 9.0, 'XL2'] = 'XS'
```

```python
# Phân loại năm lớp 12
df.loc[df['TBM3'] < 5, 'XL3'] = 'Y'
df.loc[(df['TBM3'] >= 5) & (df['TBM3'] < 6.5), 'XL3'] = 'TB'
df.loc[(df['TBM3'] >= 6.5) & (df['TBM3'] < 8.0), 'XL3'] = 'K'
df.loc[(df['TBM3'] >= 8.0) & (df['TBM3'] < 9.0), 'XL3'] = 'G'
df.loc[df['TBM3'] >= 9.0, 'XL3'] = 'XS'
```

9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization

```python
# Convert TBM1 (10-point scale) to US_TBM1 (4-point scale)
min_tbm1 = df['TBM1'].min()
max_tbm1 = df['TBM1'].max()
df['US_TBM1'] = (df['TBM1'] - min_tbm1) / (max_tbm1 - min_tbm1) * 4
```

```python
# Convert TBM2 (10-point scale) to US_TBM2 (4-point scale)
min_tbm2 = df['TBM2'].min()
max_tbm2 = df['TBM2'].max()
df['US_TBM2'] = (df['TBM2'] - min_tbm2) / (max_tbm2 - min_tbm2) * 4
```

```python
# Convert TBM3 (10-point scale) to US_TBM3 (4-point scale)
min_tbm3 = df['TBM3'].min()
max_tbm3 = df['TBM3'].max()
df['US_TBM3'] = (df['TBM3'] - min_tbm3) / (max_tbm3 - min_tbm3) * 4
```

10. Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị '1') và rớt ( giá trị '0') vào các khối dựa trên điểm DH1, DH2 và DH3 như sau
- Với khối A, A1 nếu [(DH1*2 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, 
ngược lại là rớt
- Với khối B nếu [(DH1 + DH2*2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược 
lại là rớt
- Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt

```python
df['KQXT'] = 0 
```

```python
# Tính KQXT cho khối A, A1
df.loc[(df['DH1']*2 + df['DH2'] + df['DH3'])/4 >= 5.0, 'KQXT'] = 1

# Tính KQXT cho khối B 
df.loc[(df['DH1'] + df['DH2']*2 + df['DH3'])/4 >= 5.0, 'KQXT'] = 1

# Tính KQXT cho các khối khác
df.loc[(df['DH1'] + df['DH2'] + df['DH3'])/3 >= 5.0, 'KQXT'] = 1
```
