2. Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1) 
Gợi ý - 
Trực quan hóa biểu đồ Box-plot, histogram và  QQ-plot  theo phân nhóm là giá trị của 'phanlopT1'.

```python
groups = lab2_df.groupby('phanlopt1')
```

```python
# Vẽ biểu đồ Box-plot
fig, ax = plt.subplots(figsize=(12, 6))
groups.boxplot(column='T1', ax=ax)
ax.set_title('Box-plot of T1 by phanlopT1')
ax.set_xlabel('phanlopT1')
ax.set_ylabel('T1')
plt.show()
```

```python
# Vẽ biểu đồ Histogram
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i, (name, group) in enumerate(groups):
    row = i // 2
    col = i % 2
    group['T1'].hist(ax=axes[row, col], bins=20, edgecolor='white')
    axes[row, col].set_title(f"Histogram of T1 for phanlopT1 = {name}")
plt.suptitle('Histograms of T1 by phanlopT1 groups')
plt.show()
```

```python
import seaborn as sns
# Vẽ biểu đồ QQ-plot
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i, (name, group) in enumerate(groups):
    row = i // 2
    col = i % 2
    stats.probplot(group['T1'], dist="norm", plot=axes[row, col])
    axes[row, col].set_title(f"QQ-plot of T1 for phanlopT1 = {name}")
plt.suptitle('QQ-plots of T1 by phanlopT1 groups')
plt.show()
```

3. Hãy khảo sát tương quan giữa biến DH1 theo biến T1 
Gợi ý - - 
Nhận xét giá trị Covariance hoặc Correlation 
Vẽ biểu đồ Scatter thể hiện liên hệ của biến phụ thuộc DH1 theo biến độc lập T1 

```python
# Tính toán covariance và correlation
cov = lab2_df['DH1'].cov(lab2_df['T1'])
corr = lab2_df['DH1'].corr(lab2_df['T1'])

print(f"Covariance between DH1 and T1: {cov:.2f}")
print(f"Correlation between DH1 and T1: {corr:.2f}")
```

```python
# Vẽ biểu đồ scatter 
plt.figure(figsize=(8, 6))
plt.scatter(lab2_df['T1'], lab2_df['DH1'])
plt.xlabel('T1')
plt.ylabel('DH1')
plt.title('Scatter plot of DH1 vs T1')
plt.show()
```
