cau 1: tạo Dataframe

```python
import pandas as pd
```

```python
test_df = pd.DataFrame({
    'Name': ['Alice','Bob','Charlie','David','Eva','Frank','Grace','Hannah','Ivan','Jack','Kelly','Liam','Mona','Nina','Oscar'],
    'Age': [25,30,35,28,22,45,34,31,27,29,33,40,26,32,36],
    'Salary': [50000,60000,70000,55000,52000,80000,72000,68000,61000,59000,63000,77000,53000,66000,75000]
})
```

```python
test_df
```

cau 2: hiển thị thông tin về Dataframe

```python
test_df.info()
```

cau 3: Lọc các hàng trong Dataframe có 'Age' lớn hơn 28

```python
cau3 = test_df[test_df['Age'] > 28]
cau3
```

cau 4: tính giá trị trung bình của cột 'Salary'

```python
average_salary = test_df['Salary'].mean()
print("Average Salary:", average_salary)
```

cau 5: nhóm dữ liệu theo cột 'Age' và tính tổng 'Salary cho mỗi nhóm

```python
salary_sum_by_age = test_df.groupby('Age')['Salary'].sum()
salary_sum_by_age
```

cau 6: sắp xếp Dataframe theo cột 'Salary' giảm dần

```python
df_sorted = test_df.sort_values(by='Salary', ascending=False)
df_sorted
```

cau 11: Kiểm tra xem có giá trị NaN nào trong Dataframe không

```python
test_df.isna().sum()
```

cau 12 thay thế các giá trị của cột 'Age' lớn hơn 30 bằng giá trị trung bình của cột đó

```python
test_df['Age_2'] = test_df['Age'].apply(lambda x: test_df['Age'].mean() if x > 30 else x)
test_df
```

cau 13: chuẩn hóa của cột 'Age' về khoảng giá trị từ 0 đến 1

```python
test_df['Age_normalized'] = (test_df['Age'] - test_df['Age'].min()) / (test_df['Age'].max() - test_df['Age'].min())
test_df
```

cau 14

```python
def age_group(age):
    if age < 30:
        return 'Young'
    elif age < 60:
        return 'Middle-aged'
    else:
        return 'Old'

test_df['Age_group'] = test_df['Age'].apply(age_group)
test_df
```

cau 15

```python
test_df['Salary_pct_change'] = test_df['Salary'].pct_change() * 100
print("DataFrame with Percentage Change of 'Salary':")
test_df
```

cau 16

```python
df_no_duplicates = test_df.drop_duplicates(subset='Name', keep='first')
df_no_duplicates
```
