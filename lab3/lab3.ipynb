Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. 
Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn "kg"

Vấn đề 7: "Missing values", vấn đề này xảy ra tại các cột"Age", "Weight" và "Heart Rate". Thiếu dữ liệu (dữ liệu không đầy đủ) là vấn đề xảy ra nhiều trong các nguồn dữ liệu do nhiều nguyên nhân chủ quan lẫn khách quan.Có một vài giải pháp để xử lý vấn đề này, chủ yếu dựa trên kinh nghiệm và nghiệp vụ về tập dữ liệu đó. Một số giải pháp đưa đề xuất từ chuyên gia như sau:
a. Deletion: Remove records with missing values
b. Dummy substitution: Replace missing values with a dummy but valid value: 
e.g.: 0 for numerical values.
c. Mean substitution: Replace the missing values with the mean.
d. Frequent substitution: Replace the missing values with the most frequent item.
e. Improve the data collector: Your business folk will talk to the clients and inform them about why it is worth fixing the problem with the data collector.

Yêu cầu:
- Thống kê thông tin dữ liệu thiếu trên từng biến Age và Weight 
- Yêu cầu xử lý dữ liệu thiếu như sau: Nếu dòng nào có Age hoặc Weight có dữ liệu thì phần Age hoặc Weight được tính như bên dưới, nếu thiếu cả 2 thông tin thì xóa dòng
- Age: Giá trị thay thế là mean của các giá trị trong cột Age

Vấn đề 8: 'một cột chứa quá nhiều thông tin cần được phân rã', như trong bài toán này ta thấy header 'm0006' chứa các nội dung bao gồm: m -> male, 1218 ~ 12-18 (mm-dd). Còn giá trị thì là kết quả huyết áp.

Chúng ta sẽ tách nội dung của cột này ra làm 3 cột sau: PulseRate : giá trị huyết áp, Sex: giới tính ( m: male, f: female) và time: thời gian (tháng-ngày) như sau


```python
lab3_df = pd.melt(lab3_df, id_vars=['Id', 'Firstname', 'Lastname', 'Age', 'Weight'], 
                  value_name='PulseRate', var_name='sex_and_time').sort_values(['Id', 'Firstname', 'Lastname', 'Age', 'Weight'])
tmp_df=lab3_df['sex_and_time'].str.extract("(\D)(\d+)(\d{2})",expand=True)
tmp_df.columns=['Sex','hours_lower','hours_upper']
tmp_df['Time']= tmp_df['hours_lower']+"-"+tmp_df['hours_upper']
lab3_df=pd.concat([lab3_df,tmp_df],axis=1)
lab3_df=lab3_df.drop(['sex_and_time','hours_lower','hours_upper'],axis=1)
lab3_df
```

11. Hãy khảo sát tỉ lệ dữ liệu thiếu trên biến huyết áp. Dữ liệu bị thiếu thì hãy xử lý bằng 
phương pháp sau
- Thay thế bằng giá trị trung bình liền trước và liền sau của người đó. Nếu không được thì dùng 2)
- Thay thế bằng giá trị trung bình 2 giá liền trước của người đó. Nếu không được thì dùng 3)
- Thay thế bằng giá trị trung bình 2 giá liền sau của người đó. Nếu không được thì dùng 4)
- Trung bình của các giá trị huyết áp của người đó. Nếu không được thì dùng 5).
- Trung bình của các giá trị huyết áp của nhóm giới tính. Nếu không được thì dùng 6)
- Trung bình của các giá trị dữ liệu. Nếu không được thì thay bằng mức ổn định 
trong y học.

```python
lab3_df['PulseRate'].isnull().mean()*100
```

```python
# Chuyển đổi cột PulseRate sang kiểu số, thay thế các giá trị không thể chuyển đổi bằng NaN
lab3_df['PulseRate'] = pd.to_numeric(lab3_df['PulseRate'], errors='coerce')
```

```python
def fill_missing_pulse_rate(df):
    for idx, row in df.iterrows():
        if pd.isnull(row['PulseRate']):
            # Lấy chỉ số hàng hiện tại
            current_index = df.index.get_loc(idx)
            
            # Giá trị liền trước và liền sau
            prev_value = df.iloc[current_index - 1]['PulseRate'] if current_index - 1 >= 0 else np.nan
            next_value = df.iloc[current_index + 1]['PulseRate'] if current_index + 1 < len(df) else np.nan
            
            # Thay thế bằng giá trị trung bình của liền trước và liền sau nếu cả hai đều không bị thiếu
            if not pd.isnull(prev_value) and not pd.isnull(next_value):
                df.at[idx, 'PulseRate'] = (prev_value + next_value) / 2
            
            # Thay thế bằng giá trị trung bình của 2 giá trị liền trước
            elif current_index - 2 >= 0:
                prev_values = df.iloc[current_index - 2:current_index]['PulseRate']
                if not prev_values.isnull().any():
                    df.at[idx, 'PulseRate'] = prev_values.mean()
            
            # Thay thế bằng giá trị trung bình của 2 giá trị liền sau
            elif current_index + 2 < len(df):
                next_values = df.iloc[current_index + 1:current_index + 3]['PulseRate']
                if not next_values.isnull().any():
                    df.at[idx, 'PulseRate'] = next_values.mean()
            
            # Thay thế bằng giá trị trung bình của tất cả giá trị huyết áp của người đó
            else:
                person_values = df[(df['Id'] == row['Id']) & (~df['PulseRate'].isnull())]['PulseRate']
                if not person_values.empty:
                    df.at[idx, 'PulseRate'] = person_values.mean()
                
                # Thay thế bằng giá trị trung bình của các giá trị huyết áp trong nhóm giới tính
                elif not pd.isnull(row['Sex']):
                    sex_values = df[(df['Sex'] == row['Sex']) & (~df['PulseRate'].isnull())]['PulseRate']
                    if not sex_values.empty:
                        df.at[idx, 'PulseRate'] = sex_values.mean()
                
                # Thay thế bằng giá trị trung bình của tất cả các giá trị huyết áp
                else:
                    overall_mean = df['PulseRate'].mean()
                    if not pd.isnull(overall_mean):
                        df.at[idx, 'PulseRate'] = overall_mean
                    else:
                        # Thay thế bằng mức ổn định trong y học
                        df.at[idx, 'PulseRate'] = 120
    return df

# Áp dụng hàm để thay thế giá trị thiếu trong cột PulseRate
lab3_df = fill_missing_pulse_rate(lab3_df)
lab3_df
```
